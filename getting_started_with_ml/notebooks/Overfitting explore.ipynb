```python
num_columns = list(training_set.corr()['SalePrice'].sort_values(ascending=False)[1:].index)

# remove any with NaN values
num_na_columns = list(training_set.columns[training_set.isna().any()])

num_columns = [x for x in num_columns if x not in num_na_columns]
```

```python
results = pd.DataFrame({'n':range(1, 40), 'test_error':np.repeat(0, 39), 'train_error':np.repeat(0, 39)})
```

```python
training_set_sub = training_set
```

```python
for i in range(1, 40):
    # Create features data frame and train model    
    training_columns = list(num_columns[0:i])
    predictor = linear_model.LinearRegression()
    predictor.fit(training_set_sub[training_columns], training_set_sub['SalePrice'])
    train_predictions = predictor.predict(training_set_sub[training_columns])
    train_error = metrics.mean_absolute_error(training_set_sub['SalePrice'], train_predictions)
    test_predictions = predictor.predict(test_data[training_columns])
    test_error = metrics.mean_absolute_error(test_data['SalePrice'], test_predictions)
    
    # Evaluate model
    results['test_error'][i - 1] = test_error
    results['train_error'][i - 1] = train_error
```

```python
results = pd.DataFrame({'max_depth':range(1, 40), 'test_error':np.repeat(0, 39), 'train_error':np.repeat(0, 39)})
```

```python
for i in range(1, 40):
    # Create features data frame and train model    
    training_columns = list(num_columns[0:39])
    predictor = tree.DecisionTreeRegressor(max_depth = i)
    predictor.fit(training_set[training_columns], training_set['SalePrice'])
    train_predictions = predictor.predict(training_set[training_columns])
    train_error = metrics.mean_absolute_error(training_set['SalePrice'], train_predictions)
    test_predictions = predictor.predict(test_data[training_columns])
    test_error = metrics.mean_absolute_error(test_data['SalePrice'], test_predictions)
    
    # Evaluate model
    results['test_error'][i - 1] = test_error
    results['train_error'][i - 1] = train_error
```
